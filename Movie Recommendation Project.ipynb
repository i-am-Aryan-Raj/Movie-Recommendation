## Need of using NLP

#### Stemming needs to be prerformed, to cut down repetitive words

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    list = []
    for i in text.split():    # split nhi likha to letters bhi alag ho jaenge
        list.append(ps.stem(i))

    return ' '.join(list)
```

```python
stem(final_df['tags'][0])
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
final_df
```

# The final function

```python
final_df[final_df['title'] == 'Avatar'].index
```

```python
final_df[final_df['title'] == 'Avatar'].index[0]
```

```python
def recommend(movie):
    index = final_df[final_df['title'] == movie].index[0]
    distances = similarity[index]
    recommendations = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]
    for i in recommendations:
        print(final_df['title'][i[0]])
```

```python
recommend('Brave')
```

```python
recommend('The Avengers')
```
